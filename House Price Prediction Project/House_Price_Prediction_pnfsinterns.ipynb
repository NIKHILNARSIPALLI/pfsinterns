```python
# let's get all columns of the df train, where dtype is object and make it to a list

categorical_columns = train.select_dtypes(include=['object']).columns.tolist()
categorical_columns
```

```python
# Since, there are no missing values
# Let's one-hot encode the data

train_clean = pd.get_dummies(train, columns = categorical_columns, dtype=int)
```

# test data pre-processing

```python
# Now let's proceed to clean the test data

test.head()
```

```python
#let's check for null values show only those columns whose value is greater than 0

test.isnull().sum()[test.isnull().sum()>0]
```

```python
# First let's drop the columns that we don't use


test.drop(['Alley','Id','PoolQC','Fence','MiscFeature','MasVnrType','FireplaceQu','LotFrontage'],axis=1,inplace=True)
```

```python
# let's get the percentage of null values

(test.isnull().sum()[test.isnull().sum()>0]/len(test))*100
```

```python
# let's fill the missing values


for column in test.columns:
    if test[column].dtype == 'object':
        test[column] = test[column].fillna(test[column].mode()[0])
    else:
        test[column] = test[column].fillna(test[column].mean())
```

```python
# Let's check the missing values again

test.isnull().sum()[test.isnull().sum()>0]
```

```python
# Since there are no more missing values
# let's one-hot encode the data

test_clean = pd.get_dummies(test, columns = categorical_columns, dtype=int)
```

```python
test.head()
```

# Model Building

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
# Since the model is small let's not create train test split rather we directly use Logistic Regression

X = train_clean.drop('SalePrice', axis=1)
y = train_clean['SalePrice']
```

```python
# Let's get the columns in train_X that are not there in test_clean

missing_columns = [col for col in X.columns if col not in test_clean.columns]
missing_columns
```

```python
# Let's remove these columns from X

X = X.drop(missing_columns, axis=1)
```

```python
# Now that our columsn have been matched with the test dataset
# Let's standardized the features

scaler = StandardScaler()
X_train= scaler.fit_transform(X)
sacled_test = scaler.transform(test_clean)
```

```python
# Lets create a model

model = LogisticRegression()
model.fit(X_train, y)
```

```python
# Make predictions
y_pred = model.predict(X_train)

# Evaluate the model
accuracy = accuracy_score(y, y_pred)

# Print the evaluation metrics
print(f'Accuracy: {accuracy}')
```

```python
# Right now the data has not been trained on unseed data

# Now, let's predict the test data set

test_pred = model.predict(sacled_test)

test_pred
```

```python
# Let's add test_pred to test data frame

test['SalePrice'] = test_pred
```
